```
import pathlib
import requests

year = 2024
day = 9

url = f"https://adventofcode.com/{year}/day/{day}/input"

with open("cookie.txt", "r") as file:
    session = file.read()

headers = {"cookie": f"session={session}"}

pathlib.Path("inputs").mkdir(parents=True, exist_ok=True)

with open(f"inputs/day_{str(day).zfill(2)}_input.txt", "wb") as file:
    file.write(requests.get(url, headers=headers).content)
```

```
inp = lines[0].strip()

# inp = "2333133121414131402"
disk = []

for i, c in enumerate(inp):
    if i % 2 == 0:
        for j in range(int(c)):
            disk.append(i // 2)
    else:
        for j in range(int(c)):
            disk.append(".")

disk_copy = disk.copy()

print(disk)

while True:
    try:
        first_gap = disk_copy.index(".")
    except ValueError:
        break

    last_number = disk_copy.pop()
    if last_number != ".":
        disk_copy[first_gap] = last_number

print(disk_copy)
```

```
ans = 0
for i, c in enumerate(disk_copy):
    ans += i * c
print(ans)
```

```
inp = lines[0].strip()

disk = []

for i, c in enumerate(inp):
    if i % 2 == 0:
        disk.append((i // 2, int(c)))
    else:
        disk.append((".", int(c)))

disk_copy = disk.copy()

while len(disk_copy) > 0:
    to_move = disk_copy.pop()

    if to_move[0] == ".":
        continue

    new_spot = None
    old_spot = disk.index(to_move)

    for i, d in enumerate(disk):
        if d[0] == "." and d[1] >= to_move[1] and i <= old_spot:
            new_spot = i
            break

    if new_spot is not None:

        old = disk[old_spot]
        old_length = old[1]

        new_spot_length = disk[new_spot][1]

        disk[new_spot] = to_move

        disk[old_spot] = (".", old_length)

        if new_spot_length > to_move[1]:
            disk.insert(new_spot + 1, (".", new_spot_length - old_length))

ans = 0
disk_loc = 0
for i, c in enumerate(disk):
    if c[0] != ".":
        subt = 0
        for j in range(c[1]):
            subt += (disk_loc + j) * c[0]
        ans += subt
        # print(disk_loc, c[0], subt)
    disk_loc += c[1]
print(ans)
```
